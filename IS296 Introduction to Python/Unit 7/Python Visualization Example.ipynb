### Matplotlib Example 

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Arrays for X and Y values 
x = []
y = []
for i in range(10):
    # Numbers may not make much sense for grpahics 
    x.append(np.random.randint(100))
    y.append(np.random.randint(100))
```

```python
print(x)
print(y)
```

```python
# Non Random Array
x_non_rand = []
y_non_rand = []
for i in range(10):
    x_non_rand.append(i)
    y_non_rand.append(i)
```

```python
# The most basic plot a line plot
plt.plot(x, y)
plt.show()
```

```python
# for a line non-random values are better!
plt.plot(x_non_rand, y_non_rand)
plt.show()
```

```python
# Scatter plot example
plt.plot(x, y, 'o')
plt.show()
```

```python
plt.scatter(x,y)
plt.show
```

```python
# Bar grpah example 
counts = [5, 10, 11, 6, 2]
grades = ['A', 'B', 'C', 'D', 'F']
plt.bar(grades, counts)
plt.show()
```

```python
# Working with Images
import matplotlib.image as mpimg # for managing images
img = mpimg.imread("puppy.jpg") # Read in the image directly form file
plt.imshow(img)
```

### 

```python
# Working with Matplotlib styples and subplots 
fig, ax = plt.subplots()
ax.step(x_non_rand, y_non_rand, linewidth=2.5)
plt.show()
```
